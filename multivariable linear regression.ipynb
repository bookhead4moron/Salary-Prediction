```python
df = pd.read_excel(r'/content/gdrive/My Drive/Colab Notebooks/hiring.xlsx')
df
```

```python
import math
median_test = df.testscore.median()
```

```python
df.testscore = df.testscore.fillna(median_test)
df.experience = df.experience.fillna(0.0)
```

```python
df
```

```python
%matplotlib inline
plt.xlabel('experience')
plt.ylabel('salary')
plt.scatter(df.experience,df.salary,color='red',marker='+')
```

```python
plt.xlabel('test_score')
plt.ylabel('salary')
plt.scatter(df.testscore,df.salary,color='green')
```

```python
plt.xlabel('interview score')
plt.ylabel('salary')
plt.scatter(df.interview,df.salary,color='blue',marker='*')
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['experience','testscore','interview']],df.salary)
```
